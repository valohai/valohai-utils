```python
import valohai
```

```python
params = {
    "iterations": 500,
    "learning-rate": 0.001,
}

inputs = {"my-optional-input": ""}
```

```python
valohai.prepare(
    step="train",
    default_parameters=params,
    default_inputs=inputs,
    image="valohai/pypermill",
)
```

```python
from time import sleep
import json

for i in range(valohai.parameters("iterations").value):
    print(
        json.dumps(
            {
                "accuracy": round(0.5 + (i / 126.4), 3),
                "loss": round((50 - i) * (50 - i) / 100.1, 3),
            }
        )
    )
    sleep(0.33)
```
